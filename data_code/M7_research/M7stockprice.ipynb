```python
import yfinance as yf
import pandas as pd
import datetime

# M7 종목 리스트
tickers = ["TSLA", "NVDA", "AMZN", "MSFT", "AAPL", "GOOGL", "META"]

# 오늘 날짜와 5년 전 날짜 설정
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=5*365)  # 대략 5년치 데이터

# 여러 종목 데이터를 한 번에 다운로드 (종가만 선택)
close_prices = yf.download(tickers, start=start_date, end=end_date)['Close']

# CSV 파일로 저장
file_path = "M7_close_prices.csv"
close_prices.to_csv(file_path, index=True)

print(f"✅ Saved M7 종가 데이터 to {file_path}")
```

```python
import yfinance as yf
import pandas as pd
import os
import datetime

# M7 종목 리스트
tickers = ["TSLA", "NVDA", "AMZN", "MSFT", "AAPL", "GOOGL", "META"]

# 오늘 날짜와 5년 전 날짜 설정
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=5*365)  # 대략 5년치 데이터

# 저장할 폴더 생성
output_folder = "M7_stock_data"
os.makedirs(output_folder, exist_ok=True)

# 모든 종목의 데이터를 한 번에 다운로드 (더 빠르고 효율적)
stock_data = yf.download(tickers, start=start_date, end=end_date)['Close']

# 종목별로 개별 CSV 파일 저장 (최적화된 방식)
{ stock_data[[ticker]].to_csv(os.path.join(output_folder, f"{ticker}_close_price.csv")) for ticker in stock_data.columns }

print("✅ All M7 종가 CSV files saved successfully!")
```
