```python
import pandas as pd
import yfinance as yf

# M7 ê¸°ì—…ë“¤ì˜ í‹°ì»¤ ì‹¬ë³¼
tickers = {
    'Apple': 'AAPL',
    'Microsoft': 'MSFT',
    'Alphabet': 'GOOGL',
    'Amazon': 'AMZN',
    'Meta': 'META',
    'NVIDIA': 'NVDA',
    'Tesla': 'TSLA'
}

# ì‹œì‘ ë° ì¢…ë£Œ ë‚ ì§œ ì„¤ì •
start_date = '2019-01-01'  # 50ì¼, 200ì¼ ê³„ì‚°ì„ ìœ„í•´ ë” ì˜¤ë˜ëœ ë‚ ì§œ ì„¤ì •
end_date = '2025-03-13'

# ê²°ê³¼ ì €ì¥ìš© ë°ì´í„°í”„ë ˆì„ ë¦¬ìŠ¤íŠ¸
dataframes = []

# ê° ê¸°ì—…ì˜ ì£¼ê°€ ë°ì´í„° ë‹¤ìš´ë¡œë“œ ë° ì´ë™í‰ê·  ê³„ì‚°
for company, ticker in tickers.items():
    print(f"ğŸ“¥ {company} ({ticker}) ë°ì´í„° ë‹¤ìš´ë¡œë“œ ì¤‘...")

    # ë°ì´í„° ë‹¤ìš´ë¡œë“œ
    df = yf.download(ticker, start=start_date, end=end_date)

    # ì´ë™í‰ê·  ê³„ì‚° (SMA_50, SMA_200)
    df['SMA_50'] = df['Close'].rolling(window=50).mean()
    df['SMA_200'] = df['Close'].rolling(window=200).mean()

    # MultiIndex ì»¬ëŸ¼ ë¬¸ì œ í•´ê²° (ì»¬ëŸ¼ëª… ì •ë¦¬)
    df.columns = [col if isinstance(col, str) else col[0] for col in df.columns]

    # 'Ticker' ì»¬ëŸ¼ ì¶”ê°€ (ê¸°ì—… ì‹ë³„ì„ ìœ„í•´)
    df.insert(0, 'Ticker', ticker)

    # NaNì´ ì•„ë‹Œ ê°’ë§Œ í•„í„°ë§ (SMA_50, SMA_200ì´ ì •ìƒ ê³„ì‚°ëœ ë°ì´í„°)
    df_filtered = df.dropna(subset=['SMA_50', 'SMA_200'])

    # ê²°ê³¼ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    dataframes.append(df_filtered)

    # ìµœê·¼ 5ì¼ ë°ì´í„° ì¶œë ¥ (í™•ì¸ìš©)
    print(f"\nğŸ“Š {company} ({ticker}) - ìµœê·¼ 5ì¼ ë°ì´í„°:")
    print(df_filtered[['Ticker', 'Close', 'SMA_50', 'SMA_200']].tail(5))
    print("-" * 80)

# ëª¨ë“  ê¸°ì—… ë°ì´í„° í•©ì¹˜ê¸°
final_df = pd.concat(dataframes)

# CSV íŒŒì¼ë¡œ ì €ì¥
csv_filename = "M7_stock_data.csv"
final_df.to_csv(csv_filename, index=True)  # index=True â†’ ë‚ ì§œ í¬í•¨

print(f"\nâœ… ë°ì´í„° ì €ì¥ ì™„ë£Œ: {csv_filename}")
```

```python
import pandas as pd
import yfinance as yf
import pandas_ta as ta  # ëª¨ë©˜í…€ ì§€í‘œ ê³„ì‚° ë¼ì´ë¸ŒëŸ¬ë¦¬

# M7 ê¸°ì—…ë“¤ì˜ í‹°ì»¤ ì‹¬ë³¼
tickers = {
    'Apple': 'AAPL',
    'Microsoft': 'MSFT',
    'Alphabet': 'GOOGL',
    'Amazon': 'AMZN',
    'Meta': 'META',
    'NVIDIA': 'NVDA',
    'Tesla': 'TSLA'
}

# ì‹œì‘ ë° ì¢…ë£Œ ë‚ ì§œ ì„¤ì •
start_date = '2019-01-01'  # 50ì¼, 200ì¼ ì´ë™í‰ê·  ê³„ì‚°ì„ ìœ„í•´ ë” ì˜¤ë˜ëœ ë‚ ì§œ ì„¤ì •
end_date = '2025-03-13'

# ê²°ê³¼ ì €ì¥ìš© ë°ì´í„°í”„ë ˆì„ ë¦¬ìŠ¤íŠ¸
dataframes = []

# ê° ê¸°ì—…ì˜ ì£¼ê°€ ë°ì´í„° ë‹¤ìš´ë¡œë“œ ë° ì´ë™í‰ê·  + ëª¨ë©˜í…€ ì§€í‘œ ê³„ì‚°
for company, ticker in tickers.items():
    print(f"ğŸ“¥ {company} ({ticker}) ë°ì´í„° ë‹¤ìš´ë¡œë“œ ì¤‘...")

    # ë°ì´í„° ë‹¤ìš´ë¡œë“œ
    df = yf.download(ticker, start=start_date, end=end_date)

    # âœ… ì´ë™í‰ê·  ê³„ì‚° (SMA_50, SMA_200)
    df['SMA_50'] = df['Close'].rolling(window=50).mean()
    df['SMA_200'] = df['Close'].rolling(window=200).mean()

    # âœ… ëª¨ë©˜í…€ ì§€í‘œ ì¶”ê°€
    df['Momentum_10'] = df['Close'] - df['Close'].shift(10)  # 10ì¼ ì „ ê°€ê²©ê³¼ ë¹„êµ
    df['ROC_10'] = ((df['Close'] - df['Close'].shift(10)) / df['Close'].shift(10)) * 100  # ë³€ë™ë¥ 
    df['RSI_14'] = ta.rsi(df['Close'], length=14)  # RSI (14ì¼)

    # MACD ê³„ì‚° (ê¸°ë³¸ ì„¤ì •: 12, 26, 9)
    macd_result = ta.macd(df['Close'])
    if macd_result is not None:
        df['MACD'] = macd_result['MACD_12_26_9']
        df['MACD_Signal'] = macd_result['MACDs_12_26_9']
    else:
        df['MACD'], df['MACD_Signal'] = None, None
        print(f"âš ï¸ {company} ({ticker})ì˜ MACD ê°’ì´ Noneì…ë‹ˆë‹¤.")

    # ìŠ¤í† ìºìŠ¤í‹± ì˜¤ì‹¤ë ˆì´í„° (Stochastic)
    stoch_result = ta.stoch(df['High'], df['Low'], df['Close'])
    if stoch_result is not None:
        df['Stoch_K'] = stoch_result['STOCHk_14_3_3']
    else:
        df['Stoch_K'] = None
        print(f"âš ï¸ {company} ({ticker})ì˜ Stochastic ê°’ì´ Noneì…ë‹ˆë‹¤.")

    # âœ… MultiIndex ì»¬ëŸ¼ ë¬¸ì œ í•´ê²° (ì»¬ëŸ¼ëª… ì •ë¦¬)
    df.columns = [col if isinstance(col, str) else col[0] for col in df.columns]

    # âœ… 'Ticker' ì»¬ëŸ¼ ì¶”ê°€ (ê¸°ì—… ì‹ë³„ì„ ìœ„í•´)
    df.insert(0, 'Ticker', ticker)

    # âœ… NaNì´ ì•„ë‹Œ ê°’ë§Œ í•„í„°ë§ (ì´ë™í‰ê· , ëª¨ë©˜í…€ ì§€í‘œê°€ ì •ìƒ ê³„ì‚°ëœ ë°ì´í„°)
    df_filtered = df.dropna(subset=['SMA_50', 'SMA_200', 'Momentum_10', 'ROC_10', 'RSI_14', 'MACD', 'Stoch_K'])

    # âœ… ê²°ê³¼ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    dataframes.append(df_filtered)

    # âœ… ìµœê·¼ 5ì¼ ë°ì´í„° ì¶œë ¥ (í™•ì¸ìš©)
    print(f"\nğŸ“Š {company} ({ticker}) - ìµœê·¼ 5ì¼ ë°ì´í„°:")
    print(df_filtered[['Ticker', 'Close', 'SMA_50', 'SMA_200', 'Momentum_10', 'ROC_10', 'RSI_14', 'MACD', 'Stoch_K']].tail(5))
    print("-" * 80)

# âœ… ëª¨ë“  ê¸°ì—… ë°ì´í„° í•©ì¹˜ê¸°
final_df = pd.concat(dataframes)

# âœ… CSV íŒŒì¼ë¡œ ì €ì¥
csv_filename = "M7_stock_momentum_data.csv"
final_df.to_csv(csv_filename, index=True)  # index=True â†’ ë‚ ì§œ í¬í•¨

print(f"\nâœ… ë°ì´í„° ì €ì¥ ì™„ë£Œ: {csv_filename}")
```

```python
import pandas as pd
import yfinance as yf
import pandas_ta as ta  # pandas_ta ë¼ì´ë¸ŒëŸ¬ë¦¬ ì‚¬ìš©

# M7 ê¸°ì—…ë“¤ì˜ í‹°ì»¤ ì‹¬ë³¼
tickers = {
    'Apple': 'AAPL',
    'Microsoft': 'MSFT',
    'Alphabet': 'GOOGL',
    'Amazon': 'AMZN',
    'Meta': 'META',
    'NVIDIA': 'NVDA',
    'Tesla': 'TSLA'
}

# ì‹œì‘ ë° ì¢…ë£Œ ë‚ ì§œ ì„¤ì •
start_date = '2019-01-01'
end_date = '2025-03-13'

# ê²°ê³¼ ì €ì¥ìš© ë°ì´í„°í”„ë ˆì„ ë¦¬ìŠ¤íŠ¸
dataframes = []

# âœ… ê° ê¸°ì—… ë°ì´í„° ë‹¤ìš´ë¡œë“œ ë° ì²˜ë¦¬
for company, ticker in tickers.items():
    print(f"ğŸ“¥ {company} ({ticker}) ë°ì´í„° ë‹¤ìš´ë¡œë“œ ì¤‘...")

    # ë°ì´í„° ë‹¤ìš´ë¡œë“œ
    df = yf.download(ticker, start=start_date, end=end_date)

    # âœ… MultiIndex ì»¬ëŸ¼ í•´ê²°
    df.columns = ['_'.join(col) if isinstance(col, tuple) else col for col in df.columns]

    # âœ… ì»¬ëŸ¼ ì´ë¦„ ë³€ê²½ (Close_AAPL â†’ Close ë“±)
    df = df.rename(columns={f'Close_{ticker}': 'Close', f'High_{ticker}': 'High',
                            f'Low_{ticker}': 'Low', f'Open_{ticker}': 'Open',
                            f'Volume_{ticker}': 'Volume'})

    # âœ… ë°ì´í„° í™•ì¸
    print(df.head())  # ë°ì´í„° ìƒ˜í”Œ ì¶œë ¥
    print("\nğŸ“Œ [Close ì»¬ëŸ¼ íƒ€ì… í™•ì¸]:", df['Close'].dtype)

    # âœ… ì´ë™í‰ê·  ê³„ì‚° (SMA_50, SMA_200)
    df['SMA_50'] = df['Close'].rolling(window=50).mean()
    df['SMA_200'] = df['Close'].rolling(window=200).mean()

    # âœ… ëª¨ë©˜í…€ ì§€í‘œ ì¶”ê°€
    df['Momentum_10'] = df['Close'] - df['Close'].shift(10)
    df['ROC_10'] = ((df['Close'] - df['Close'].shift(10)) / df['Close'].shift(10)) * 100
    df['RSI_14'] = ta.rsi(df['Close'], length=14)

    # âœ… MACD ê³„ì‚°
    macd_result = ta.macd(df['Close'])
    if macd_result is not None:
        df['MACD'] = macd_result['MACD_12_26_9']
        df['MACD_Signal'] = macd_result['MACDs_12_26_9']
    else:
        df['MACD'], df['MACD_Signal'] = 0, 0
        print(f"âš ï¸ {company} ({ticker})ì˜ MACD ê°’ì´ Noneì´ë¯€ë¡œ 0ìœ¼ë¡œ ì±„ì›ë‹ˆë‹¤.")

    # âœ… Stochastic ê³„ì‚°
    stoch_result = ta.stoch(df['High'], df['Low'], df['Close'])
    if stoch_result is not None:
        df['Stoch_K'] = stoch_result['STOCHk_14_3_3']
    else:
        df['Stoch_K'] = 0
        print(f"âš ï¸ {company} ({ticker})ì˜ Stochastic ê°’ì´ Noneì´ë¯€ë¡œ 0ìœ¼ë¡œ ì±„ì›ë‹ˆë‹¤.")

    # âœ… 'Ticker' ì»¬ëŸ¼ ì¶”ê°€ (ê¸°ì—… ì‹ë³„ì„ ìœ„í•´)
    df.insert(0, 'Ticker', ticker)

    # âœ… ê²°ì¸¡ê°’(NaN) ì œê±°
    df_filtered = df.dropna(subset=['SMA_50', 'SMA_200', 'Momentum_10', 'ROC_10', 'RSI_14'])

    # âœ… ê²°ê³¼ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    dataframes.append(df_filtered)

    # âœ… ìµœê·¼ 5ì¼ ë°ì´í„° ì¶œë ¥ (í™•ì¸ìš©)
    print(f"\nğŸ“Š {company} ({ticker}) - ìµœê·¼ 5ì¼ ë°ì´í„°:")
    print(df_filtered[['Ticker', 'Close', 'SMA_50', 'SMA_200', 'Momentum_10', 'ROC_10', 'RSI_14', 'MACD', 'Stoch_K']].tail(5))
    print("-" * 80)

# âœ… ëª¨ë“  ê¸°ì—… ë°ì´í„° í•©ì¹˜ê¸°
final_df = pd.concat(dataframes)

# âœ… CSV íŒŒì¼ë¡œ ì €ì¥
csv_filename = "M7_stock_momentum_data_fixed.csv"
final_df.to_csv(csv_filename, index=True)

print(f"\nâœ… ë°ì´í„° ì €ì¥ ì™„ë£Œ: {csv_filename}")
```
