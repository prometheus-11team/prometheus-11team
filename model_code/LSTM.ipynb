LSTM 모델 학습용 텐서 형태 `(samples, time_steps, features)`로 가공

````python
import numpy as np
from sklearn.preprocessing import StandardScaler

# 🔧 설정
df_final = merged_df_final.drop(columns=["Company"])  # Company(기업명) 열 제거, 티커만 남김
window_size = 30  # 시퀀스 길이
feature_cols = [col for col in df_final.columns if col not in ['Date', 'Ticker', 'Close', 'Open', 'High', 'Low']]  # 제외할 열 지정

# 결과 저장
X_list, y_list = [], []

# 티커별 반복
for ticker in df_final["Ticker"].unique():
    df_t = df_final[df_final["Ticker"] == ticker].sort_values("Date").reset_index(drop=True)

    # 스케일링 (각 티커별로)
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(df_t[feature_cols])
    
    # 종가 수익률 계산
    close = df_t["Close"].values
    returns = (np.roll(close, -1) - close) / close  # 다음날 수익률
    returns = returns[:-1]  # 마지막은 NaN
    labels = (returns > 0).astype(int)  # 상승이면 1, 하락이면 0

    # 슬라이딩 윈도우 구성
    for i in range(len(df_t) - window_size - 1):
        x = scaled_features[i:i+window_size]
        y = labels[i+window_size]
        X_list.append(x)
        y_list.append(y)

# 최종 텐서로 변환
X = np.array(X_list)
y = np.array(y_list)

# 출력 정보
print("✅ X shape:", X.shape)  # (샘플 수, 시퀀스 길이, 피처 수)
print("✅ y shape:", y.shape)  # (샘플 수,)
````

```

labels = (returns > 0).astype(int)              # 수익률이 양수면 1 (상승), 아니면 0 (하락)
```
---> 따라서 y는 "30일 시퀀스 다음 날의 주가가 상승했는지 여부 (1/0)"

````python
np.unique(y, return_counts=True)
````

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# LSTM 모델 정의
model_tf = Sequential([
    LSTM(64, input_shape=(window_size, X.shape[2]), return_sequences=True),
    Dropout(0.2),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')  # 이진 분류를 위한 sigmoid 활성화 함수
])

# 모델 컴파일
model_tf.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 모델 학습
history = model_tf.fit(X, y, epochs=20, batch_size=32, validation_split=0.2)

# 학습 결과 출력
print("✅ 모델 학습 완료")
````

````python
 # 학습 결과 시각화
plt.figure(figsize=(12, 4))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
````

````python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
````

##### 1. Pytorch Dataset 클래스 정의

````python
class StockDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
````

##### 2. LSTM 모델 정의

````python
class LSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_layers=2, dropout=0.2):
        super(LSTM, self).__init__()
        self.lstm = nn.LSTM(input_size = input_dim, 
                            hidden_size = hidden_dim, 
                            num_layers=num_layers, 
                            dropout=dropout, 
                            batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        out, _ = self.lstm(x)  # out: (batch, seq_len, hidden)
        out = out[:, -1, :]    # 마지막 time step만 사용
        out = self.fc(out)
        return self.sigmoid(out).squeeze(1)  # shape: (batch,)
````

##### 3. 학습 루프

````python
from sklearn.model_selection import train_test_split

# 데이터 분할
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, shuffle=False)

# DataLoader 생성
train_dataset = StockDataset(X_train, y_train)
val_dataset = StockDataset(X_val, y_val)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64)

# 모델 초기화
input_dim = X.shape[2]
model = LSTM(input_dim=input_dim)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# 손실 함수와 옵티마이저
criterion = nn.BCELoss()  # Binary Cross-Entropy
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# 학습 루프
epochs = 10
for epoch in range(epochs):
    model.train()
    total_loss = 0

    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)

        optimizer.zero_grad()
        preds = model(X_batch)
        loss = criterion(preds, y_batch)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs} | Train Loss: {total_loss / len(train_loader):.4f}")
````

````python
from sklearn.metrics import accuracy_score

model.eval()
y_true, y_pred = [], []

with torch.no_grad():
    for X_batch, y_batch in val_loader:
        X_batch = X_batch.to(device)
        preds = model(X_batch)
        preds_class = (preds.cpu().numpy() > 0.5).astype(int)
        y_true.extend(y_batch.numpy())
        y_pred.extend(preds_class)

acc = accuracy_score(y_true, y_pred)
print(f"✅ 검증 정확도: {acc:.4f}")
````

````python
from sklearn.preprocessing import StandardScaler

# 1. 전처리: feature_cols 정의 및 scaler 학습
feature_cols = [col for col in df_final.columns if col not in ['Date', 'Ticker', 'Close', 'Open', 'High', 'Low']]  # 제외할 열 지정

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_final[feature_cols])
df_final[feature_cols] = scaled_features  # 스케일 적용

# 2. 정렬 보장
df_final = df_final.sort_values(["Ticker", "Date"]).reset_index(drop=True)

# 3. 시퀀스 데이터 생성
X_all = []
tickers_all = []
dates_all = []
y_true_all = []

window_size = 30

for ticker in df_final["Ticker"].unique():
    df_ticker = df_final[df_final["Ticker"] == ticker]
    features = df_ticker[feature_cols].values

    if len(features) <= window_size:
        continue  # 너무 짧은 시계열은 스킵

    for i in range(len(features) - window_size):
        X_all.append(features[i:i+window_size])
        y_true_all.append(df_ticker.iloc[i + window_size]["y_true"])
        tickers_all.append(ticker)
        dates_all.append(df_ticker.iloc[i + window_size]["Date"])

X_all = np.array(X_all)  # (samples, time_steps, features)

# 4. 모델 예측
y_prob_all = model_tf.predict(X_all, verbose=0)  # shape: (samples, 1)
y_pred_all = (y_prob_all > 0.5).astype(int).flatten()

# 5. 결과 저장
result_df = pd.DataFrame({
    "Date": dates_all,
    "Ticker": tickers_all,
    "y_true": y_true_all,
    "y_pred_tf": y_pred_all,
    "y_prob_tf": y_prob_all.flatten()
})
result_df
````
